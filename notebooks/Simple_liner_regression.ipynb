```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Importing the dataset
dataset = pd.read_csv('../data/car data.csv')

# Check the first few rows to understand the data
print(dataset.head())

# Basic data info (check for nulls or datatype issues)
print(dataset.info())
print(dataset.describe())

# Handle missing values if any (drop or fill)
dataset = dataset.dropna()  # Simple approach, remove rows with missing data

# Select features and target variable
# Assuming 'TV' is the feature and 'Sales' is the target, change if different
X = dataset[['TV']].values  # Feature matrix should be 2D
y = dataset['Sales'].values

# Splitting the dataset into the Training set and Test set (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

# Training the Simple Linear Regression model on the Training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# Evaluate the model
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, regressor.predict(X_train), color='red', label='Regression line')
plt.title('Sales vs TV (Training set)')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color='green', label='Test data')
plt.plot(X_train, regressor.predict(X_train), color='red', label='Regression line')
plt.title('Sales vs TV (Test set)')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
